# Working With Some Multiple Plots

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
t1 = np.arange(0,5,.2);
t2 = np.arange(0,5,.4);
plt.figure(1)
#  Above Line Of Code Create A Squre For You In Which You Can  Divide The Space For Your Fig.
plt.subplot(2,1,1)
# 2,1,1 = > create A Squre in diffrent Parts Linke 2 = Create Row, 1 Colums in Squre another 1 is for Space Where You Want To Work 
plt.grid()
plt.plot(t1,f(t1),'b--')
plt.subplot(2,1,2)
plt.plot(t2,np.cos(2*np.pi*t2),'r--')
plt.show

```
